### OLS slope estimates converge faster than intercept

```python
import numpy as np
T = np.arange(100)
u = np.random.normal(0,1,100)
y = 10+0.5*T+u

def ols(y,X):
    X = np.c_[np.ones((100,1)), T]
    Y = y
    XX = np.dot(X.T,X)
    XY = np.dot(X.T,Y)
    return np.dot(np.linalg.inv(XX), XY)

beta0s, beta1s = [], []
N = 100
for i in range(N):
    u = np.random.normal(0,1,100)
    y = 10+0.5*T+u
    beta0, beta1 = ols(y,np.c_[np.ones((100,1)), T])
    beta0s.append(1/np.sqrt(N)*(beta0-10))
    beta1s.append(1/np.sqrt(N)*(beta1-0.5))

import plotly.io as pio
pio.renderers.default = 'iframe'
import plotly.express as px
px.histogram(x=[beta0s,beta1s])
```

### Cointegration

```python
T = 1000
ρ = 0.5
u = np.random.normal(0,1,T)
v = np.random.normal(0,1,T)
x = np.zeros(T)
y = np.zeros(T)
for t in range(T):
    x[t] = x[t-1]+v[t]
    y[t] = ρ*y[t-1]+u[t]
px.line(y=[x,y],x=np.arange(T))
```

### Stationary, Explosive and Unit root processes

```python
T = 1000
u = np.random.normal(0,1,T)
y1 = np.zeros(T)
y2 = np.zeros(T)
y3 = np.zeros(T)
for t in range(T):
    y1[t] = 0.5*y1[t-1]+u[t]
    y2[t] = 1*y2[t-1]+u[t]
    y3[t] = 1.02*y3[t-1]+u[t]

px.line(y=[y1[0:100],y2[0:100],y3[0:100]],x=np.arange(100))
```

### Brownian Motion

```python
# increments
T = 10000
u = np.random.choice([-1,1],T)
v = np.random.choice([-1,1],T)  
x = np.cumsum(u)
y = np.cumsum(v)
px.line(y=[x,y])
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
