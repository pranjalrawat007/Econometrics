### Cyclical Model

```python
import numpy as np
import pandas as pd
import plotly.express as px

T = np.arange(100)

def model(a, ω, θ, T):
    Y = 2*a*np.cos(ω*T+θ)
    return Y
```

```python
# Phase Shift
px.line(y=[model(1, 0.2, 0, T), model(1, 0.2, 0.4, T)],x=T)
```

```python
# Frequency
px.line(y=[model(1, 0.2, 0, T), model(1, 0.5, 0, T)],x=T)
```

```python
# Amplitude
px.line(y=[model(1, 0.2, 0, T), model(2, 0.2, 0, T)],x=T)
```

### Shrumpeter's Model: Additive Model with multiple cycles

```python
T = np.arange(1000)

def model2(a, ω, θ, T):
    Y = np.zeros(T.shape)
    for i in range(len(a)):
        Y += model(a[i], ω[i], θ[i], T)
    return Y

a = [5,2,1]
ω = [0.03, 0.20, 0.47]
θ = [0,0,0]
px.line(y=[model2(a, ω, θ, T)],x=T)
```

```python
# Spectrum
px.scatter(y=[i**2 for i in a],x=ω)    
```
