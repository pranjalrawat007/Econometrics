### OLS slope estimates converge faster than intercept

```python
import numpy as np
T = np.arange(100)
u = np.random.normal(0,1,100)
y = 10+0.5*T+u

def ols(y,X):
    X = np.c_[np.ones((100,1)), T]
    Y = y
    XX = np.dot(X.T,X)
    XY = np.dot(X.T,Y)
    return np.dot(np.linalg.inv(XX), XY)

beta0s, beta1s = [], []
for i in range(100):
    u = np.random.normal(0,1,100)
    y = 10+0.5*T+u
    beta0, beta1 = ols(y,X)
    
    beta0s.append(1/10*(beta0-10))
    beta1s.append(1/10*(beta1-0.5))

import plotly.express as px
px.histogram(x=[beta0s,beta1s])
```
