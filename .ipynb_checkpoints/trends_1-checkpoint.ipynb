### Simple Trends in Macroeconomics Variables

```python
import pandas_datareader.data as web    # pandas 0.19.x and later
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import datetime
start = datetime.datetime(1960, 1, 1)
end = datetime.datetime(2023, 1, 27)
df = pd.DataFrame([])
df['gdp'] = np.log(web.DataReader("GDPC1", "fred", start, end))
df['con'] = np.log(web.DataReader("PCEPI", "fred", start, end))
df['inv'] = np.log(web.DataReader("GPDIC1", "fred", start, end))
df['p'] = np.log(web.DataReader("GDPDEF", "fred", start, end))
df['t'] = np.array(range(df.shape[0]))
df.head()
```

```python
df.tail()
```

```python
smf.ols(formula='gdp ~ t', data=df).fit().summary()
```

```python
smf.ols(formula='con ~ t', data=df).fit().summary()
```

```python
smf.ols(formula='inv ~ t', data=df).fit().summary()
```

### annual growth rate of y = 0.0072*4 = 2.8%

```python
df['gdpd'] = df.gdp-np.array(smf.ols(formula='gdp ~ t', data=df).fit().predict())
df['cd'] = df.con-np.array(smf.ols(formula='con ~ t', data=df).fit().predict())
df['id'] = df.inv-np.array(smf.ols(formula='inv ~ t', data=df).fit().predict())
```

```python
def autocorr(x, df):
    print(x)
    x = df[x]
    print(np.corrcoef(x.iloc[0:-1], x.iloc[1:])[0,1])
    print(np.corrcoef(x.iloc[0:-2], x.iloc[2:])[0,1])
    print(np.corrcoef(x.iloc[0:-3], x.iloc[3:])[0,1])
    print(np.corrcoef(x.iloc[0:-4], x.iloc[4:])[0,1])
  
```

```python
autocorr('gdpd', df)
```

```python
autocorr('cd', df)
```

```python
autocorr('id', df)
```

```python
df['gdpg'] = df['gdp']-df['gdp'].shift()
df['cong'] = df['con']-df['con'].shift()
df['invg'] = df['inv']-df['inv'].shift()
df['inf'] = df['p']-df['p'].shift()
```

```python
def analysis(start, end):
    start = datetime.datetime(start, 1, 1)
    end = datetime.datetime(end, 1, 27)
    df = pd.DataFrame([])
    df['gdp'] = np.log(web.DataReader("GDPC1", "fred", start, end))
    df['con'] = np.log(web.DataReader("PCEPI", "fred", start, end))
    df['inv'] = np.log(web.DataReader("GPDIC1", "fred", start, end))
    df['p'] = np.log(web.DataReader("GDPDEF", "fred", start, end))
    df = df.dropna()
    df['t'] = np.array(range(df.shape[0]))
    print(smf.ols(formula='gdp ~ t', data=df).fit().summary())
    print(smf.ols(formula='con ~ t', data=df).fit().summary())
    print(smf.ols(formula='inv ~ t', data=df).fit().summary())
    df['gdpd'] = df.gdp-np.array(smf.ols(formula='gdp ~ t', data=df).fit().predict())
    df['cond'] = df.con-np.array(smf.ols(formula='con ~ t', data=df).fit().predict())
    df['invd'] = df.inv-np.array(smf.ols(formula='inv ~ t', data=df).fit().predict())
    print(df.head)

    autocorr('gdpd', df)
    autocorr('cond', df)
    autocorr('invd', df)
    df['gdpg'] = df['gdp']-df['gdp'].shift()
    df['cong'] = df['con']-df['con'].shift()
    df['invg'] = df['inv']-df['inv'].shift()
    df['inf'] = df['p']-df['p'].shift()
    df = df.dropna()
    autocorr('gdpg', df)
    autocorr('cong', df)
    autocorr('invg', df)
    autocorr('inf', df)
    return df
```

```python
df = analysis(1952, 1970)
```

```python
df = analysis(1970, 2023)
```
